```python
#cleaning and pre-processing of the twitter dataset
def clean_text(text):
    #remove the links
    text = re.sub(r"http\S+","", text)
    #remove special characters, numbers, punctuations  
    text = re.sub("[^a-zA-Z#]"," ",text) 
    #replace "$&@*#" with a common word vulgar
    text = text.replace("$&@*#", "vulgar")
    # convert text to lowercase 
    text = text.lower() 
    
    
    return text
```

```python
train_df_upsampled['tweet'] = train_df_upsampled['tweet'].apply(lambda x: clean_text(x))
#test_df['tweet'] = test_df['tweet'].apply(lambda x: clean_text(x))
```

```python
#function to remove the # character
def remove_hashtag(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = input_txt.replace('#','')
        
    return input_txt  
    
```

```python
#function to remove the @word tags
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt  
```

```python
train_df_upsampled['tweet'] = np.vectorize(remove_pattern)(train_df_upsampled['tweet'], "@[\w]*")
train_df_upsampled['tweet'] = np.vectorize(remove_hashtag)(train_df_upsampled['tweet'], "#[\w]*")
```

```python
train_df_upsampled['tweet'] = train_df_upsampled['tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>1]))
#test_df['tweet'] = test_df['tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>1]))
```

```python
#next approach will be to remove the stopwords and analyze the words
def freq_words(x, terms = 30): 
    
    all_words = ' '.join([text for text in x]) 
    all_words = all_words.split() 
    fdist = nltk.FreqDist(all_words) 
    words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) 
  
  # selecting top 20 most frequent words 
    d = words_df.nlargest(columns="count", n = terms) 
  
  # visualize words and frequencies
    plt.figure(figsize=(12,15)) 
    ax = sns.barplot(data=d, x= "count", y = "word") 
    ax.set(ylabel = 'Word') 
    plt.show()
  

freq_words(tokenized_train_df['tweet'], 100)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# function to remove stopwords
def remove_stopwords(text):
    no_stopword_text = [w for w in text.split() if not w in stop_words]
    return ' '.join(no_stopword_text)

tokenized_train_df['tweet'] = tokenized_train_df['tweet'].apply(lambda x: remove_stopwords(x))
```

```python
#Using a pre trained Glove model for extracting out feature vectors of 100 dimensions.
import itertools
from glove import Corpus, Glove
sentences = list(itertools.islice(tokenized_train_df['tweet'],None))
new_lines =[]
for sentence in sentences:
   new_lines.append(sentence.split(' '))

corpus = Corpus()
corpus.fit(new_lines, window=10)
glove = Glove(no_components=100, learning_rate=0.1)
glove_model = glove.fit(corpus.matrix, epochs= 64, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)
# window , no_components , epochs of training of glove model and the learning can be changed and the change in results can be seen 
```

```python
glove.save('glove.model')
print(glove.word_vectors[glove.dictionary['samsung']])
```

```python
train_glove =[]
for sentence in new_lines:
    train_glove.append(np.mean([glove.word_vectors[glove.dictionary[sentence[i]]] for i in range(len(sentence))],axis =0))
```

```python
# Oversample and plot imbalanced dataset with SMOTE
# It can be seen in the below plot how new feature vectors have been added for imbalanved data
from imblearn.over_sampling import SMOTE
# transform the dataset
oversample = SMOTE()
X, y = oversample.fit_resample(train_glove,train_df_upsampled['label'])
# summarize the new class distribution
counter = Counter(y)
print(counter)
# scatter plot of examples by class label
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.legend()
plt.show()
```

```python
xtrain_glove.shape
```

```python
xtrain_glove = xtrain_glove.reshape(9430,100,1)
```

```python
import time
import keras
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, Dropout
lstm_out=100
model = Sequential()
model.add(LSTM(lstm_out, dropout=0.1,input_shape=(100,1)))
model.add(Dense(2,activation='softmax'))
start = time.time()
model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics = ['accuracy'])
print("Compilation Time : ", time.time() - start)
print(model.summary())
```

```python
xvalid_glove.shape
```

```python
xvalid_glove=xvalid_glove.reshape(2358,100,1)
```

```python
y_train_dummy = pd.get_dummies(ytrain)
y_valid_dummy = pd.get_dummies(yvalid)
```

```python
#Using appropriate callbacks like EarlyStopping , Learning Rate schedules can also be used.
from keras.callbacks import EarlyStopping
import tensorflow as tf
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=6)
history = model.fit(xtrain_glove,y_train_dummy, epochs =32, batch_size = 32,callbacks=[callback],
    validation_data=(xvalid_glove, y_valid_dummy),verbose = 1)
```

```python
#Plotting the history
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
score,acc = model.evaluate(xvalid_glove,y_valid_dummy, verbose = 2, batch_size = 32)
print("score: %.2f" % (score))
print("acc: %.2f" % (acc))
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
y_pred1 = model.predict(xvalid_glove,batch_size =16, verbose =1)
y_pred = np.argmax(y_pred1, axis=1)
y_pred= pd.get_dummies(y_pred)
# Print f1, precision, and recall scores
print(precision_score(y_valid_dummy, y_pred , average="weighted"))
print(recall_score(y_valid_dummy, y_pred , average="weighted"))
print(f1_score(y_valid_dummy, y_pred , average="weighted"))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_valid_dummy, y_pred))
```
